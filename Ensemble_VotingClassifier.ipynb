```python
df_cat = df.select_dtypes("object")
df_num = df.select_dtypes(['float64','int64']).iloc[:,:-1]
y = df['HeartDisease']
```

```python
encoder = LabelEncoder()
df_encoded = df_cat.apply(encoder.fit_transform)
```

```python
X_con = np.concatenate([df_encoded.values , df_num.values] , axis=1 )
X = pd.DataFrame(X_con)
```

```python
scale = MinMaxScaler()
x = scale.fit_transform(X)
```

```python
x_train , x_test , Y_train , y_test = train_test_split(x , y , test_size= 0.2 , random_state= 123)
```

```python
clf1 = LogisticRegression( random_state=1)
clf2 = RandomForestClassifier(n_estimators=50, random_state=1)
clf3 = GaussianNB()
```

```python
clf = VotingClassifier([clf1,clf2,clf3])
```
